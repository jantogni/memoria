```python
# Generating VECM synthetic data
mu, sigma = 0, 10 # mean and standard deviation
n = 10000      # number of samples
l = 2   # number of variables
p = 2   # number of lags
epsilon = np.random.normal(mu, sigma, n*l)
epsilon=epsilon.reshape(l,n)
phi_1 = [[-0.2, 0.1], [0.5, 0.2]]
phi_2 = [[0.8, 0.7], [-0.4, 0.6]]
# VAR model 
y = np.zeros((l,n))
for i in range(2,n):
    y[:,i] = np.dot(phi_1,y[:,i-1]) + np.dot(phi_2,y[:,i-2]) + epsilon[:,i]
y = y - np.min(y) + 2*sigma   #  Avoiding negative numbers
todays_date = datetime.datetime.now().date()   # Converting to a dataframe object
index = pd.date_range(todays_date-datetime.timedelta(10), periods=n, freq='D')
df = pd.DataFrame(index=index, data={'col1': y[0,:], 'col2': y[1,:]} )
df.plot()
#df.to_csv('synthetic-data.csv')
```

This data is generated from the following VAR(2) model:

$
\mathbf{y}_t = 
\begin{bmatrix} 
-0.2 & 0.1 \\
0.5 & 0.2
\end{bmatrix} \mathbf{y}_{t-1} +
\begin{bmatrix} 
0.8 & 0.7 \\
-0.4 & 0.6
\end{bmatrix} \mathbf{y}_{t-2}+
\epsilon_t
$ 

with
$
\Sigma = 
\begin{bmatrix} 
100 & 0 \\
0   & 100
\end{bmatrix}
$
and $\mathbf{y}_{-1} = \mathbf{y}_{-1} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$


The VECM(2) representation with cointegration rank 1 is:

$
\Delta \mathbf{y}_t = 
\begin{bmatrix} 
-0.4 \\
0.1 
\end{bmatrix} 
\begin{bmatrix} 
1 & -2 
\end{bmatrix}
\mathbf{y}_{t-1} -
\begin{bmatrix} 
0.8 & 0.7 \\
-0.4 & 0.6
\end{bmatrix} \Delta \mathbf{y}_{t-1}+
\epsilon_t
$
