```python
def MAPE_LS(y_true, y_pred):
    """ Mean absolute percentage error

    Parameters
    ----------
    y_true:
    y_pred:

    Return
    MAPE: Mean absolute percentage error
    """
    try:
        l = y_true.shape[1]
    except:
        y_true.shape += (1,)  # avoiding (l,) shape problem
        y_pred.shape += (1,)  # avoiding (l,) shape problem
        l = 1
    mape = np.zeros(l)

    for i in range(l):
        indexes = y_true[:, i].nonzero()
        actual = y_true[:, i][indexes]
        forecast = y_pred[:, i][indexes]
        #print "Stock", i
        #print np.mean(np.abs(actual - forecast))
        #print np.sum(np.abs(actual))
        mape[i] = 100 * np.sum(np.abs(actual - forecast))/np.sum(np.abs(actual))

    return mape
```
