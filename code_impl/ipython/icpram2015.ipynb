```python
import sys
sys.path.insert(0, '../python/')
import classes as cl
import algorithms as alg
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.tsa.vector_ar as vector_ar
```

```python
def OARIMA(Y, order, L, it):
    """
    Sliding window ARIMA
    
    Parameters
    ----------
    y:      time series
            Pandas Dataframe
    order:  [p,d,q] ARIMA parameters
            List
    L:      Sliding window size
            Integer
    it:     Number of iterations
    """
    l = Y.shape[1]
    Y_true = np.zeros([it, l],dtype=np.float32)
    Y_pred = np.zeros([it, l],dtype=np.float32)
    DY_true = np.zeros([it, l],dtype=np.float32)
    DY_pred = np.zeros([it, l],dtype=np.float32)
    mape = np.zeros([it, l],dtype=np.float32)
    mae = np.zeros([it, l],dtype=np.float32)
    rmse = np.zeros([it, l],dtype=np.float32)
    stats = cl.stats()
    
    for i in range(it):
        for j in range(l):
            y = Y.iloc[i:i+L,j]
            Y_true[i,j] = Y.iloc[i+L,j]
            try:
                model = ARIMA(y,order,freq='D')
                fit = model.fit()
                Y_pred[i,j] = fit.forecast(steps=1)[0][0]
                DY_true[i,j] = Y_true[i,j] - Y.iloc[i+L-1,j]
                DY_pred[i,j] = Y_pred[i,j] - Y.iloc[i+L-1,j]
                y_true = y[1:]
                y_pred = model.predict(fit.params)+y[:-1]
                stats = cl.stats(y_true.as_matrix(), y_pred.as_matrix())
                mape[i,j], mae[i,j], rmse[i,j] = stats.mape, stats.mae, stats.rmse
            except LinAlgError:
                pass
                #print "it=",i,",stock=",j
    #mape, mae, rmse = stats.MAPE(Y_true, Y_pred), stats.MAE(Y_true, Y_pred), stats.RMSE(Y_true, Y_pred)    
            
    results = cl.model_it(Y_true, Y_pred, DY_true, DY_pred, mape, mae, rmse)
    return results
```

```python
def OVECM(y, L, P, it, avg_error, r, n):
    it2, l = y.shape

    dy_true = np.zeros([it, l],dtype=np.float32)
    dy_pred = np.zeros([it, l],dtype=np.float32)
    mape = np.zeros([it, l],dtype=np.float32)
    mae = np.zeros([it, l],dtype=np.float32)
    rmse = np.zeros([it, l],dtype=np.float32)
    
    m = cl.Matrix()
    
    start_time = time.time()

    for i in range(it):
        #sys.stdout.write("\r{0}%".format((float(i + 1) / it) * 100))
        #sys.stdout.flush()
        y_i = y[i:i+L]

        if i == 0:
            #Initialization
            beta = m.get_johansen(y_i.as_matrix(), P, r)
            A, B = m.vec_matrix(y_i, P, beta.evecr)
        else:
            #Update & Out-of-sample forecast
            A, B = m.vec_matrix_online(A, B, y_i, P, beta.evecr)
            dy_true[i-1,:] = B[-1,:]
            dy_pred[i-1,:] = np.dot(A[-1,:], x)

        #OLS
        x, residuals, rank, s = np.linalg.lstsq(A, B)
        Ax = np.dot(A, x)

        #Internal mape
        y_true = y_i.as_matrix()[-n:]
        y_pred = Ax[-n:] + y_i.as_matrix()[-n-1:-1]
        
        #Stats info
        stats = cl.stats(y_true, y_pred)
        mape[i,:], mae[i,:], rmse[i,:] = stats.mape, stats.mae, stats.rmse
        avg_mape = np.average(mape[i,:])
        
        #Beta update
        #print avg_mape
        if avg_mape > avg_error:
            beta = m.get_johansen(y_i.as_matrix(), P)            
            r = beta.r
            A = m.vec_matrix_update(A, y_i, P, beta.evecr)
            x, residuals, rank, s = np.linalg.lstsq(A, B)

    oecm = cl.model_it(y[L:L+it], dy_pred + y[L-1:L+it-1], dy_true, dy_pred, mape, mae, rmse)
    return oecm

def OOVECM(y, L, P, it, r, n):
    it2, l = y.shape

    dy_true = np.zeros([it, l])
    dy_pred = np.zeros([it, l], dtype=np.float32)
    mape = np.zeros([it, l], dtype=np.float32)
    mae = np.zeros([it, l], dtype=np.float32)
    rmse = np.zeros([it, l], dtype=np.float32)

    m = cl.Matrix()
    start_time = time.time()
    for i in range(1, it):
        y_i = y[i:i + L]
        beta = m.get_johansen(y_i.as_matrix(), P, r)
        A, B = m.vec_matrix(y_i, P, beta.evecr)

        x, residuals, rank, s = np.linalg.lstsq(A, B)
        Ax = np.dot(A, x)

        dy_true[i, :] = B[-1,:]
        dy_pred[i, :] = Ax[-1,:]
        
        y_true = y_i.as_matrix()[-n:]
        y_pred = Ax[-n:] + y_i.as_matrix()[-n - 1:-1]
        
        stats = cl.stats(y_true, y_pred)
        mape[i, :], mae[i,:], rmse[i,:] = stats.mape, stats.mae, stats.rmse

    oecm = cl.model_it(y[L:L + it], dy_pred + y[L - 1:L + it - 1], dy_true, dy_pred, mape, mae, rmse)
    return oecm
```
