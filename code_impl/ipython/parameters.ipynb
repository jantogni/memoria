```python
def akaike_cmp(residuals, nx, L, nlag):
    nobs, neqs = residuals.shape
    k = neqs * nlag + nx + 1
    T = nobs * neqs
    square = residuals**2
    RSS = sum(square)
    Loglikelihood = -(T/2)*(1+log(2*pi)+log(RSS/T))
    AIC_m = -2*Loglikelihood/T + 2*k/T
    AIC_rss = T * log(RSS/T) + 2*k
    return AIC_m, AIC_rss
```
